```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('/content/drive/MyDrive/데이터분석/brazilEcommerce/olist_customers_dataset.csv')
reviews = pd.read_csv('/content/drive/MyDrive/데이터분석/brazilEcommerce/olist_order_reviews_dataset.csv')
payments = pd.read_csv('/content/drive/MyDrive/데이터분석/brazilEcommerce/olist_order_payments_dataset.csv')
orders = pd.read_csv('/content/drive/MyDrive/데이터분석/brazilEcommerce/olist_orders_dataset.csv')
items = pd.read_csv('/content/drive/MyDrive/데이터분석/brazilEcommerce/olist_order_items_dataset.csv')
products = pd.read_csv('/content/drive/MyDrive/데이터분석/brazilEcommerce/olist_products_dataset.csv')
```

```python
reviews = reviews[['order_id', 'review_id', 'review_score']]
```

```python
order_review = pd.merge(orders, reviews)
order_review.head()
```

```python
order_review_over = order_review.set_index('customer_id')
```

```python
condition = (order_review['customer_id'].value_counts() != 1)
order_review_over = order_review_over[condition].reset_index()
```

```python
order_review_over[['customer_id', 'order_id', 'review_score']].groupby('customer_id').mean()
```

```python
order_review_over[['customer_id', 'order_id', 'review_score']].groupby('customer_id').mean().mean()
```

```python
condition2 = (order_review['customer_id'].value_counts() == 1)
order_review_once = order_review.set_index('customer_id')
order_review_once = order_review_once[condition2].reset_index()
order_review_once[['customer_id', 'order_id', 'review_score']].groupby('customer_id').mean()
```
