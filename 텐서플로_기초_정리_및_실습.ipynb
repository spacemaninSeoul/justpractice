# 2. 딥러닝으로 풀어보는 수학문제(Linear Regression)

```python
키 = 170
신발 = 260
```

```python
#신발 = 키 * a + b
```

```python
a = tf.Variable(0.1)
b = tf.Variable(0.2)
```

```python
# a, b 경사하강법으로 구하기
opt = tf.keras.optimizers.Adam(learning_rate=0.1) # gradient를 알아서 스마트하게 바꿔줌
```

```python
# opt.minimize(손실함수, var_list=[a, b])  var_list : 경사하강법으로 업데이트할 weight Variable 목록
```

```python
def 손실함수():
  예측값 = 키 * a + b
  return tf.square(260 - 예측값)
```

```python
opt.minimize(손실함수, var_list=[a, b]) # 경사하강 1번 해줌
```

```python
for i in range(300):
  opt.minimize(손실함수, var_list=[a, b])
```

```python
print(a.numpy(), b.numpy())
```

# 3. 대학원 합격 예측 AI 만들기



```python
import tensorflow as tf
```

```python
# 1. 딥러닝 model 디자인하기
# Sequential 쓰면 신경망 레이어들 쉽게 만들어줌. Dense() 히든레이어에서의 노드 개수. 마지막 레이어는 하나의 결과를 가져야하기에 (1). 확률 한 개를 예측할 때도 1개
# 레이어에 activation functuin 넣기 (sigmoid, tanh, relu, softmax, leakyRelu)
# 마지막 레이어는 예측결과를 뱉어야 (0~1 사이의 확률)
model = tf.keras.models.Sequential([
                          tf.keras.layers.Dense(64, activation= 'tanh'),
                          tf.keras.layers.Dense(128, activation= 'tanh'),
                          tf.keras.layers.Dense(1, activation= 'sigmoid') # 0~1 사이의 확률을 뱉고싶으면 sigmoid
])
```

```python
# 2. model compile 하기
# optimizer는 w1 <= w1 - (å * ∂E/∂w1  <- 이 값을 알맞게 조정해줌)
# 0과 1 사이에서의 확률, 분류 문제를 예측할때 로스 함수는 binary_crossentropy를 쓸 수 있다.
model.compile(optimizer='adam' , loss='binary_crossentropy' , metrics=['accuracy'])
```
