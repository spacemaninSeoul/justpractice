```python
# 1번 문제
중간고사 = [34, 30, 26, 40, 33, 15, 31, 21, 17, 40]
기말고사 = [45, 48, 25, 50, 50, 28, 39, 33, 47, 42]
과제 = [5, 10, 8, 10, 10, 7, 7, 9, 10, 2]

총점 = []
for 중간, 기말, 과제점수 in zip(중간고사, 기말고사, 과제):
  총점.append(중간 + 기말 + 과제점수)

총점.sort(reverse=True)
print(총점)
```
